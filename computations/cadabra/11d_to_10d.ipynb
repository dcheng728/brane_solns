```cadabra2
def post_process(ex):
	sort_product(ex)
	canonicalise(ex)
	collect_terms(ex)
```

```cadabra2
# Christoffel symbols (Levi-Civita connection)
Gamma := \Gamma^{m}_{n p} = 1/2 g^{m q} (
   \partial_{n}{g_{q p}} + \partial_{p}{g_{n q}} - \partial_{q}{g_{n p}}
);
```

```cadabra2
# Riemann tensor in terms of Christoffel symbols
Riem := R^{m}_{n p q} = \partial_{p}{\Gamma^{m}_{n q}}
   - \partial_{q}{\Gamma^{m}_{n p}}
   + \Gamma^{m}_{s p} \Gamma^{s}_{n q}
   - \Gamma^{m}_{s q} \Gamma^{s}_{n p};
```

```cadabra2
# Ricci tensor: R_{nq} = R^{p}_{npq} (contract 1st and 3rd indices of Riemann)
Ric := R_{n q} = \partial_{p}{\Gamma^{p}_{n q}}
   - \partial_{q}{\Gamma^{p}_{n p}}
   + \Gamma^{p}_{s p} \Gamma^{s}_{n q}
   - \Gamma^{p}_{s q} \Gamma^{s}_{n p};
```

```cadabra2
# Ricci scalar: R = g^{nq} R_{nq}
RicSc := R = g^{n q} (
   \partial_{p}{\Gamma^{p}_{n q}}
   - \partial_{q}{\Gamma^{p}_{n p}}
   + \Gamma^{p}_{s p} \Gamma^{s}_{n q}
   - \Gamma^{p}_{s q} \Gamma^{s}_{n p}
);
```

```cadabra2
# Expand Ricci scalar in terms of metric and its derivatives
substitute(RicSc, Gamma);
distribute(RicSc);
product_rule(RicSc);
distribute(RicSc);
sort_product(RicSc);
canonicalise(RicSc);
collect_terms(RicSc);
```
