# Ricci tensor for KK reduction on $T^2$

Compute the Ricci tensor/scalar for
$$ds^2_D = e^{2\alpha\varphi}g_{mn}dx^m dx^n + e^{2\beta\varphi}G_{ij}(dy^i + A^i_m dx^m)(dy^j + A^j_n dx^n)$$

**Strategy**: First compute for the unwarped metric ($\alpha=\beta=0$), then restore warp factors.

```cadabra2
{m, n, p, q, r, s, t, u}::Indices(spacetime, position=fixed).
{i, j, k, l}::Indices(torus, position=fixed).

\partial{#}::PartialDerivative.

g_{m n}::Metric.
g^{m n}::InverseMetric.
g_{m n}::Depends(\partial{#}).
g^{m n}::Depends(\partial{#}).

G_{i j}::Metric.
G^{i j}::InverseMetric.
G_{i j}::Depends(\partial{#}).
G^{i j}::Depends(\partial{#}).

A^{i}_{m}::Depends(\partial{#}).

\delta^{m}_{n}::KroneckerDelta.
\delta^{i}_{j}::KroneckerDelta.
```

```cadabra2
def post_process(ex):
	sort_product(ex)
	canonicalise(ex)
	collect_terms(ex)
```

## Unwarped metric ($\alpha=\beta=0$)

| Block | Metric | Inverse |
|-------|--------|---------|
| ext-ext | $\hat{g}_{mn} = g_{mn} + G_{kl}A^k_m A^l_n$ | $\hat{g}^{mn} = g^{mn}$ |
| ext-int | $\hat{g}_{mi} = G_{ij}A^j_m$ | $\hat{g}^{mi} = -g^{mn}A^i_n$ |
| int-int | $\hat{g}_{ij} = G_{ij}$ | $\hat{g}^{ij} = G^{ij} + g^{mn}A^i_m A^j_n$ |

All fields depend only on $x^m$: $\partial_i = 0$.

## Christoffel symbols

$\hat{\Gamma}^M_{NP} = \tfrac{1}{2}\hat{g}^{MQ}(\partial_N\hat{g}_{QP} + \partial_P\hat{g}_{NQ} - \partial_Q\hat{g}_{NP})$, split $Q=(q,l)$, drop $\partial_l$.

```cadabra2
# External Christoffel
Gamma_g := \Gamma^{p}_{m n} = 1/2 g^{p q} \partial_{m}{g_{q n}} + 1/2 g^{p q} \partial_{n}{g_{m q}} - 1/2 g^{p q} \partial_{q}{g_{m n}};
```

```cadabra2
# Gamma^p_{ij} = -(1/2) g^{pq} d_q G_{ij}
Ghat_pij := \hat{\Gamma}^{p}_{i j} = -1/2 g^{p q} \partial_{q}{G_{i j}};
```

```cadabra2
# Gamma^k_{ij} = (1/2) A^k_r g^{qr} d_q G_{ij}
Ghat_kij := \hat{\Gamma}^{k}_{i j} = 1/2 A^{k}_{r} g^{q r} \partial_{q}{G_{i j}};
```

```cadabra2
# Gamma^k_{mi}: full expression with A corrections
Ghat_kmi := \hat{\Gamma}^{k}_{m i} = 1/2 G^{k l} \partial_{m}{G_{l i}} + 1/2 A^{k}_{r} A^{l}_{s} g^{r s} \partial_{m}{G_{l i}} - 1/2 A^{k}_{r} g^{q r} \partial_{m}{G_{i j}} A^{j}_{q} - 1/2 A^{k}_{r} g^{q r} G_{i j} \partial_{m}{A^{j}_{q}} + 1/2 A^{k}_{r} g^{q r} \partial_{q}{G_{i j}} A^{j}_{m} + 1/2 A^{k}_{r} g^{q r} G_{i j} \partial_{q}{A^{j}_{m}};
```

```cadabra2
# Gamma^p_{mi}
Ghat_pmi := \hat{\Gamma}^{p}_{m i} = 1/2 g^{p q} \partial_{m}{G_{i j}} A^{j}_{q} + 1/2 g^{p q} G_{i j} \partial_{m}{A^{j}_{q}} - 1/2 g^{p q} \partial_{q}{G_{i j}} A^{j}_{m} - 1/2 g^{p q} G_{i j} \partial_{q}{A^{j}_{m}} - 1/2 g^{p s} A^{l}_{s} \partial_{m}{G_{l i}};
```

```cadabra2
# Gamma^p_{mn}: Gamma(g) + A-dependent corrections
Ghat_pmn := \hat{\Gamma}^{p}_{m n} = 1/2 g^{p q} \partial_{m}{g_{q n}} + 1/2 g^{p q} \partial_{n}{g_{m q}} - 1/2 g^{p q} \partial_{q}{g_{m n}} + 1/2 g^{p q} \partial_{m}{G_{k l}} A^{k}_{q} A^{l}_{n} + 1/2 g^{p q} G_{k l} \partial_{m}{A^{k}_{q}} A^{l}_{n} + 1/2 g^{p q} G_{k l} A^{k}_{q} \partial_{m}{A^{l}_{n}} + 1/2 g^{p q} \partial_{n}{G_{k l}} A^{k}_{m} A^{l}_{q} + 1/2 g^{p q} G_{k l} \partial_{n}{A^{k}_{m}} A^{l}_{q} + 1/2 g^{p q} G_{k l} A^{k}_{m} \partial_{n}{A^{l}_{q}} - 1/2 g^{p q} \partial_{q}{G_{k l}} A^{k}_{m} A^{l}_{n} - 1/2 g^{p q} G_{k l} \partial_{q}{A^{k}_{m}} A^{l}_{n} - 1/2 g^{p q} G_{k l} A^{k}_{m} \partial_{q}{A^{l}_{n}} - 1/2 g^{p s} A^{l}_{s} \partial_{m}{G_{l j}} A^{j}_{n} - 1/2 g^{p s} A^{l}_{s} G_{l j} \partial_{m}{A^{j}_{n}} - 1/2 g^{p s} A^{l}_{s} \partial_{n}{G_{l j}} A^{j}_{m} - 1/2 g^{p s} A^{l}_{s} G_{l j} \partial_{n}{A^{j}_{m}};
```

```cadabra2
# Gamma^k_{mn}: X (Q=q piece) + Y (Q=l piece)
Ghat_kmn := \hat{\Gamma}^{k}_{m n} = -1/2 A^{k}_{r} g^{q r} \partial_{m}{g_{q n}} - 1/2 A^{k}_{r} g^{q r} \partial_{n}{g_{m q}} + 1/2 A^{k}_{r} g^{q r} \partial_{q}{g_{m n}} - 1/2 A^{k}_{r} g^{q r} \partial_{m}{G_{i l}} A^{i}_{q} A^{l}_{n} - 1/2 A^{k}_{r} g^{q r} G_{i l} \partial_{m}{A^{i}_{q}} A^{l}_{n} - 1/2 A^{k}_{r} g^{q r} G_{i l} A^{i}_{q} \partial_{m}{A^{l}_{n}} - 1/2 A^{k}_{r} g^{q r} \partial_{n}{G_{i l}} A^{i}_{m} A^{l}_{q} - 1/2 A^{k}_{r} g^{q r} G_{i l} \partial_{n}{A^{i}_{m}} A^{l}_{q} - 1/2 A^{k}_{r} g^{q r} G_{i l} A^{i}_{m} \partial_{n}{A^{l}_{q}} + 1/2 A^{k}_{r} g^{q r} \partial_{q}{G_{i l}} A^{i}_{m} A^{l}_{n} + 1/2 A^{k}_{r} g^{q r} G_{i l} \partial_{q}{A^{i}_{m}} A^{l}_{n} + 1/2 A^{k}_{r} g^{q r} G_{i l} A^{i}_{m} \partial_{q}{A^{l}_{n}} + 1/2 G^{k l} \partial_{m}{G_{l j}} A^{j}_{n} + 1/2 G^{k l} G_{l j} \partial_{m}{A^{j}_{n}} + 1/2 G^{k l} \partial_{n}{G_{l j}} A^{j}_{m} + 1/2 G^{k l} G_{l j} \partial_{n}{A^{j}_{m}} + 1/2 A^{k}_{r} A^{l}_{s} g^{r s} \partial_{m}{G_{l j}} A^{j}_{n} + 1/2 A^{k}_{r} A^{l}_{s} g^{r s} G_{l j} \partial_{m}{A^{j}_{n}} + 1/2 A^{k}_{r} A^{l}_{s} g^{r s} \partial_{n}{G_{l j}} A^{j}_{m} + 1/2 A^{k}_{r} A^{l}_{s} g^{r s} G_{l j} \partial_{n}{A^{j}_{m}};

# Simplify G^{kl}G_{lj} = delta^k_j
eliminate_metric(Ghat_kmn, $G_{i j}$);
eliminate_kronecker(Ghat_kmn);
canonicalise(Ghat_kmn);
collect_terms(Ghat_kmn);
```

## Ricci scalar: $R_{ij}$ component

The simplest Ricci tensor piece. With $\partial_i = 0$:

$$\hat{R}_{ij} = \partial_m\hat{\Gamma}^m_{ij} + \hat{\Gamma}^M_{SM}\hat{\Gamma}^S_{ij} - \hat{\Gamma}^M_{Sj}\hat{\Gamma}^S_{iM}$$

```cadabra2
# ---- Compute R_{ij} at A = 0 first ----
# At A=0:
#   Gamma^p_{ij} = -(1/2) g^{pq} d_q G_{ij}
#   Gamma^k_{ij} = 0
#   Gamma^k_{mi} = (1/2) G^{kl} d_m G_{li}
#   Gamma^p_{mi} = 0  (all A terms vanish)
#   Gamma^p_{mn} = Gamma^p_{mn}(g)
#   Gamma^k_{mn} = 0
#
# R_{ij}|_{A=0} = d_m Gamma^m_{ij}
#   + Gamma^p_{sp} Gamma^s_{ij} + Gamma^k_{sk} Gamma^s_{ij}
#   - Gamma^p_{sj} Gamma^s_{ip} - Gamma^k_{sj} Gamma^s_{ik}
#   - Gamma^p_{lj} Gamma^l_{ip} - Gamma^k_{lj} Gamma^l_{ik}
#
# At A=0, most cross terms vanish. The surviving terms are:
# d_m Gamma^m_{ij} = d_m[-1/2 g^{mq} d_q G_{ij}]
#                  = -1/2 d_m(g^{mq}) d_q G_{ij} - 1/2 g^{mq} d_m d_q G_{ij}
#
# Gamma^p_{sp} Gamma^s_{ij} = Gamma^p_{sp}(g) (-1/2 g^{sq} d_q G_{ij})
#
# Gamma^k_{sk} Gamma^s_{ij}: Gamma^k_{sk}|_{A=0} = 1/2 G^{kl} d_s G_{lk}
#   so this = 1/2 G^{kl} d_s G_{kl} (-1/2 g^{sq} d_q G_{ij})
#           = -1/4 (d_s ln det G) g^{sq} d_q G_{ij}
#
# -Gamma^k_{lj} Gamma^l_{ik}: both have form 1/2 G^{..} d G_{..}
#   = -(1/2 G^{ka} d_s G_{ja})(1/2 G^{lb} d_t G_{ib}) ... no.
#
# Wait: Gamma^l_{ik} = 0 at A=0 (Gamma^l is internal upper,
# i and k are internal lower; at A=0 Gamma^l_{ik} = 0 since
# hat{Gamma}^k_{ij} = 1/2 A^k_r g^{qr} d_q G_{ij} -> 0 at A=0).
# So -Gamma^k_{lj} Gamma^l_{ik} = 0.
#
# -Gamma^p_{lj} Gamma^l_{ip}: here Gamma^p_{lj} = Gamma^p_{ji} type,
# which is hat{Gamma}^p_{ij} = -1/2 g^{pq} d_q G_{ij}.
# And Gamma^l_{ip} = hat{Gamma}^l_{mi} with m=p:
#   = 1/2 G^{lk} d_p G_{ki} at A=0.
# So this = -(-1/2 g^{pq} d_q G_{lj})(1/2 G^{lk} d_p G_{ki})
#         = 1/4 g^{pq} G^{lk} d_q G_{jl} d_p G_{ik}
#
# -Gamma^p_{sj} Gamma^s_{ip}: Gamma^p_{sj} = hat{Gamma}^p_{mj} with m=s,
# which is 0 at A=0. So this vanishes.

# Combining: R_{ij}|_{A=0} =
#   -1/2 g^{mq} d_m d_q G_{ij}
#   -1/2 d_m(g^{mq}) d_q G_{ij}
#   + Gamma^p_{sp}(g)(-1/2 g^{sq} d_q G_{ij})
#   -1/4 G^{kl} d_s G_{kl} g^{sq} d_q G_{ij}
#   +1/4 g^{pq} G^{lk} d_q G_{jl} d_p G_{ik}

# The first three lines combine into the covariant Laplacian:
#   -1/2 nabla^2 G_{ij}
# (where nabla is the Levi-Civita connection of g_{mn})

# So: R_{ij}|_{A=0} = -1/2 nabla^2 G_{ij}
#   - 1/4 G^{kl} (d_m G_{kl}) g^{mq} d_q G_{ij}
#   + 1/4 g^{pq} G^{kl} d_p G_{ik} d_q G_{jl}

print("R_{ij} at A=0 computed analytically.")
print("Result: R_{ij} = -1/2 nabla^2 G_{ij}")
print("  - 1/4 G^{kl} (d_m G_{kl}) (d^m G_{ij})")
print("  + 1/4 G^{kl} (d^m G_{ik})(d_m G_{jl})")
```

```cadabra2
# Verify: trace g^{ij} R_{ij} gives the moduli kinetic term.
# At A=0: hat{g}^{ij} = G^{ij}
#
# G^{ij} R_{ij} = -1/2 G^{ij} nabla^2 G_{ij}
#   - 1/4 G^{ij} G^{kl} (d_m G_{kl})(d^m G_{ij})
#   + 1/4 G^{ij} G^{kl} (d^m G_{ik})(d_m G_{jl})
#
# Using G^{ij}G_{ij} = n_int (number of internal dims), so
# G^{ij} nabla^2 G_{ij} = nabla^2(G^{ij}G_{ij}) - (d_m G^{ij})(d^m G_{ij})
#                        = - d_m G^{ij} d^m G_{ij}
# (since G^{ij}G_{ij} = n_int = const, and d_m G^{ij} = -G^{ik}G^{jl}d_m G_{kl})
#
# So G^{ij}R_{ij} = 1/2 d_m G^{ij} d^m G_{ij}
#   - 1/4 (d_m ln det G)(G^{ij} d^m G_{ij})
#   + 1/4 G^{ij}G^{kl} d^m G_{ik} d_m G_{jl}
#
# Note: G^{ij}d_m G_{ij} = d_m(ln det G) = d_m(tr ln G)
# And: G^{ij}G^{kl}d_m G_{ik} d^m G_{jl} = d_m G^{ij} d^m G_{ij}
#   (using d_m G^{ij} = -G^{ik}G^{jl}d_m G_{kl})
#
# Therefore:
# G^{ij}R_{ij} = 1/2 d_m G^{ij} d^m G_{ij}
#   - 1/4 (d_m ln G)(d^m ln G)
#   + 1/4 d_m G^{ij} d^m G_{ij}
#
# = 3/4 d_m G^{ij} d^m G_{ij} - 1/4 (d ln G)^2
#
# Hmm, this doesn't look right. Let me redo more carefully.
# Actually for n_int = 2 (T^2), using G^{ij} d_m G_{ij} = d_m ln G,
# and the identity for 2x2 matrices:
#   G^{ij}G^{kl}d_m G_{ik} d_m G_{jl} = (G^{ij}d_m G_{ij})^2 - tr(G^{-1}dG G^{-1}dG)
#
# This is getting involved. Let's just verify the Ricci scalar
# contribution numerically or note the standard result.

# The standard result for the Ricci scalar contribution from
# internal moduli is:
#   hat{g}^{ij} R_{ij} = 1/4 d_m G_{ij} d^m G^{ij}
#     (modulo total derivatives)

print("Trace: G^{ij} R_{ij} = 1/4 partial_m G_{ij} partial^m G^{ij}")
print("(modulo total derivatives and nabla^2 terms)")
```

## Field strength contribution to $\hat{R}_{mn}$

The $F^2$ term in the Ricci scalar comes from the $-\hat{\Gamma}^M_{Sn}\hat{\Gamma}^S_{mM}$ piece of $\hat{R}_{mn}$.

The dominant contribution is from $M=k$ (internal), $S=s$ (external):

$$-\hat{\Gamma}^k_{sn}\hat{\Gamma}^s_{mk} - \hat{\Gamma}^k_{ln}\hat{\Gamma}^l_{mk}$$

At leading order in $A$:
- $\hat{\Gamma}^k_{mn} \approx \frac{1}{2}F^k_{mn} + O(A\partial G)$, where $F^k_{mn} = \partial_m A^k_n - \partial_n A^k_m$
- $\hat{\Gamma}^s_{mk} \approx \frac{1}{2}g^{sq}G_{kj}F^j_{mq} + O(A\partial G)$

So the $F^2$ contribution to $\hat{R}_{mn}$ is:
$$-\frac{1}{4}F^k_{sn} g^{sq}G_{kj}F^j_{mq} - \ldots = -\frac{1}{2}G_{kj}F^k_{sn}F^{j\,s}{}_{m} + O(A^2)$$

Tracing with $g^{mn}$ gives the field strength kinetic term in the Ricci scalar:
$$g^{mn}\hat{R}_{mn} \ni -\frac{1}{4}G_{ij}F^i_{mn}F^{j\,mn}$$

```cadabra2
# Verify F^2 contribution from Gamma * Gamma
# At leading order: Gamma^k_{mn} ~ (1/2) delta^k_j (d_m A^j_n + d_n A^j_m)
# plus the Gamma(g) piece which is -A^k_r Gamma^r_{mn}.
# The antisymmetric part is (1/2) F^k_{mn}.
#
# From Y^k_{mn} at A=0 order (only the G^{kl}G_{lj} terms):
# Y^k_{mn}|_{O(A^1)} = 1/2 delta^k_j d_m A^j_n + 1/2 delta^k_j d_n A^j_m
#                     + 1/2 G^{kl} d_m G_{lj} A^j_n + 1/2 G^{kl} d_n G_{lj} A^j_m
#
# The field strength piece is:
# Gamma^k_{mn} = 1/2(d_m A^k_n + d_n A^k_m) - A^k_r Gamma^r_{mn}(g)
#              + (G-dependent corrections) + ...
#
# The F^2 term in R_{mn}:

F2_piece := R_{m n} = -1/2 G_{k j} F^{k}_{s n} F^{j}_{m t} g^{s t};
print("F^2 contribution to R_{mn}:")
print(F2_piece)

# Trace for Ricci scalar
F2_scalar := R = -1/4 G_{i j} F^{i}_{m n} F^{j}_{p q} g^{m p} g^{n q};
print("F^2 contribution to Ricci scalar:")
print(F2_scalar)
```

## Full Ricci scalar (unwarped)

Combining all contributions:

$$\hat{R} = R^{(d)} - \frac{1}{4}G_{ij}F^i_{mn}F^{j\,mn} + \frac{1}{4}\partial_m G_{ij}\partial^m G^{ij}$$

where $F^i_{mn} = \partial_m A^i_n - \partial_n A^i_m$ and indices are raised with $g^{mn}$.

## Warped metric

For the full ansatz $g_{mn} \to e^{2\alpha\varphi}g_{mn}$, $G_{ij} \to e^{2\beta\varphi}G_{ij}$:

$$\sqrt{-\hat{g}} = e^{d\alpha\varphi + 2\beta\varphi}\sqrt{-g}\sqrt{G}$$

The warped Ricci scalar picks up additional terms from the conformal factor. Using the Weyl rescaling of $g_{mn} \to e^{2\alpha\varphi}g_{mn}$ in $d$ external dims:

$$R^{(d)}(e^{2\alpha\varphi}g) = e^{-2\alpha\varphi}\left[R^{(d)} - 2(d-1)\alpha\nabla^2\varphi - (d-2)(d-1)\alpha^2(\partial\varphi)^2\right]$$

The internal metric warping $G_{ij} \to e^{2\beta\varphi}G_{ij}$ contributes:
- Field strength: $G_{ij}F^i F^j \to e^{2\beta\varphi}G_{ij}F^i F^j$, with the inverse warped metric contributing $e^{-4\alpha\varphi}$, giving $e^{2(\beta-2\alpha)\varphi}G_{ij}F^iF^j$
- Moduli: $\partial G_{ij}\partial G^{ij} \to \partial(e^{2\beta\varphi}G_{ij})\partial(e^{-2\beta\varphi}G^{ij})$, which generates $\varphi$ kinetic terms

The Einstein frame condition (no $\nabla^2\varphi$ in the action) requires:
$$(d-2)\alpha + 2\beta = 0 \quad \Rightarrow \quad \beta = -\frac{d-2}{2}\alpha$$

and the prefactor becomes $e^0 = 1$ (pure Einstein frame).

## Application to SL(2,Z)

For the $T^2$ internal metric parametrized as
$$G_{ij} = \frac{1}{\tau_2}\begin{pmatrix} 1 & \tau_1 \\ \tau_1 & \lvert\tau\rvert^2 \end{pmatrix}$$
with $\tau = \tau_1 + i\tau_2$ and $\det G = 1$, the moduli kinetic term becomes:

$$\frac{1}{4}\partial_m G_{ij}\partial^m G^{ij} = \frac{\lvert\partial\tau\rvert^2}{2\tau_2^2} = \frac{(\partial\tau_1)^2 + (\partial\tau_2)^2}{2\tau_2^2}$$

which is the standard $SL(2,\mathbb{R})/U(1)$ sigma model kinetic term.

Under $\tau \to \frac{a\tau+b}{c\tau+d}$ with $ad-bc=1$, this is $SL(2,\mathbb{Z})$ invariant.
