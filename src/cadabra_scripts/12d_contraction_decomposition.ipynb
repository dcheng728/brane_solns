We use S= DP with rank 4, K = DP with rank2, P rank1

It is known that if rnk-4 tensor S has the same symmetries as a Riemann tensor, then

$$
\begin{align}
t_8 t_8 R^2 S^2 &= 4 R_{abcd} R_{abcd} S_{efgh} S_{efgh}
+ 16 R_{abcd} R_{abef} S_{cdgh} S_{efgh} \\
&\quad + 8 R_{abcd} R_{efgh} S_{abcd} S_{efgh}
+ 8 R_{abcd} R_{efgh} S_{abef} S_{cdgh} \\
&\quad - 64 R_{abcd} R_{abce} S_{dfgh} S_{efgh}
- 32 R_{abcd} R_{abef} S_{cegh} S_{dfgh} \\
&\quad - 64 R_{abcd} R_{aefg} S_{bhfg} S_{cdeh}
- 64 R_{abcd} R_{efgh} S_{abce} S_{dfgh} \\
&\quad - 64 R_{abcd} R_{aefg} S_{bhcd} S_{ehfg}
+ 128 R_{abcd} R_{aecf} S_{bgdh} S_{egfh} \\
&\quad + 256 R_{abcd} R_{aefg} S_{bhcf} S_{dgeh}
+ 128 R_{abcd} R_{aecf} S_{bgfh} S_{dheg} \\
&\quad + 64 R_{abcd} R_{efgh} S_{aecg} S_{bfdh}, 
\label{eq:t8R2S2}
\end{align}
$$

In particular, for $t_8 t_8 R^4$, we have

$$
\begin{align}
t_8 t_8 R^4 &= 3 \times 2^7 \Biggl\{
R_{abef} R^{fg}{}_{cd} R^{dahe} R_{gh}{}^{bc}
+ \frac{1}{2} R_{abef} R^{fgbc} R_{cdgh} R^{heda} \notag \\
&\quad - \frac{1}{2} R_{abef} R^{efbc} R_{cdgh} R^{dagh}
- \frac{1}{4} R_{abef} R^{ef}{}_{cd} R^{dagh} R_{gh}{}^{bc} \notag \\
&\quad + \frac{1}{16} R_{abef} R^{abgh} R_{ghcd} R^{cdef}
+ \frac{1}{32} (R_{abcd} R^{abcd})^2
\Biggr\}. \label{eq:A2}
\end{align}
$$

```cadabra2
def truncate(ex):
    split_index(ex, $M, m, a$, repeat=True)
    distribute(ex)
    # 1 internal index
    substitute(ex, $\hat{R}_{m n r a} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m n a r} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m a n r} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{a m n r} -> 0$, repeat=True)
    # 3 internal indices
    substitute(ex, $\hat{R}_{a b c m} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{a b m c} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{a m b c} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m a b c} -> 0$, repeat=True)
    # 2 internal in same antisym pair (O(P^2))
    substitute(ex, $\hat{R}_{a b m n} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m n a b} -> 0$, repeat=True)
    # 4 internal (O(P^2))
    substitute(ex, $\hat{R}_{a b c d} -> 0$, repeat=True)
    return ex
    
def simplify(ex):
    # simplify
    sort_product(ex)
    canonicalise(ex)
    rename_dummies(ex)
    collect_terms(ex)
    return ex
```

```cadabra2
{M,N,R,S,T,U,V,P,Q}::Indices(full,position=free).
{m,n,r,s,t,u,p,q,i,j,k,l,v,w,x,y}::Indices(spacetime,parent=full,position=free).
{a,b,c,d,e,f,g,h,a1,a2,a3,a4,a5,a6,a7,a8}::Indices(internal,parent=full,position=free).
\hat{R}_{M N R S}::RiemannTensor.
```

```cadabra2
# Line 1 (coeff 1):
ex1 := \hat{R}_{P N Q M} \hat{R}_{P R Q S} \hat{R}_{M T R U} \hat{R}_{N U S T};

# Line 2 (coeff 1/2):
ex2 := \hat{R}_{M P N Q} \hat{R}_{Q R P S} \hat{R}_{T S U R} \hat{R}_{U N T M};

# Line 3 (coeff -1/2):
ex3 := \hat{R}_{M N R S} \hat{R}_{R S N P} \hat{R}_{P Q U V} \hat{R}_{Q M U V};

# Line 4 (coeff -1/4):
ex4 := \hat{R}_{M N R S} \hat{R}_{R S P Q} \hat{R}_{Q M U V} \hat{R}_{U V N P};

# Line 5 (coeff 1/16):
ex5 := \hat{R}_{M N R S} \hat{R}_{M N P Q} \hat{R}_{P Q U V} \hat{R}_{U V R S};

# Line 6 (coeff 1/32):
ex6 := \hat{R}_{M N R S} \hat{R}_{M N R S} \hat{R}_{P Q U V} \hat{R}_{P Q U V};
```
