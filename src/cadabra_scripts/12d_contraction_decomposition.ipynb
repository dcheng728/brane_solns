```cadabra2
def simplify(ex):
    # simplify
    sort_product(ex)
    canonicalise(ex)
    rename_dummies(ex)
    collect_terms(ex)
    canonicalise(ex)
    return ex
```

```cadabra2
def truncate(ex):
    split_index(ex, $M, m, a$, repeat=True)
    distribute(ex)
    
    # ALL DOWN: 1 internal
    substitute(ex, $\hat{R}_{a m n r} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m a n r} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m n a r} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m n r a} -> 0$, repeat=True)
    # ALL DOWN: 3 internal
    substitute(ex, $\hat{R}_{a b c m} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{a b m c} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{a m b c} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m a b c} -> 0$, repeat=True)
    # ALL DOWN: 2 internal same pair
    substitute(ex, $\hat{R}_{a b m n} -> 0$, repeat=True)
    substitute(ex, $\hat{R}_{m n a b} -> 0$, repeat=True)
    # ALL DOWN: 4 internal
    substitute(ex, $\hat{R}_{a b c d} -> 0$, repeat=True)
    
    # 2 UP 2 DOWN: 1 internal
    substitute(ex, $\hat{R}^{a r}{}_{m n} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{r a}{}_{m n} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m r}{}_{a n} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m r}{}_{n a} -> 0$, repeat=True)
    
    # 2 UP 2 DOWN: 3 internal
    substitute(ex, $\hat{R}^{a b}{}_{c m} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{a b}{}_{m c} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{a m}{}_{b c} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m a}{}_{b c} -> 0$, repeat=True)
    
    # 2 UP 2 DOWN: 2 internal same pair
    substitute(ex, $\hat{R}^{a b}{}_{m n} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m n}{}_{a b} -> 0$, repeat=True)
    
    # 2 UP 2 DOWN: 4 internal
    substitute(ex, $\hat{R}^{a b}{}_{c d} -> 0$, repeat=True)
    
    # ALL UP: 1 internal
    substitute(ex, $\hat{R}^{a m n r} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m a n r} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m n a r} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m n r a} -> 0$, repeat=True)
    # ALL UP: 3 internal
    substitute(ex, $\hat{R}^{a b c m} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{a b m c} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{a m b c} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m a b c} -> 0$, repeat=True)
    # ALL UP: 2 internal same pair
    substitute(ex, $\hat{R}^{a b m n} -> 0$, repeat=True)
    substitute(ex, $\hat{R}^{m n a b} -> 0$, repeat=True)
    # ALL UP: 4 internal
    substitute(ex, $\hat{R}^{a b c d} -> 0$, repeat=True)
    
    return ex
```

```cadabra2
def lower_internal(ex):
    substitute(ex, $\hat{R}^{a m b n} -> g^{a c}g^{b d}g^{m p}g^{n q} \hat{R}_{c p d q}$, repeat=True)
    substitute(ex, $\hat{R}^{m a b n} -> -g^{a c}g^{b d}g^{m p}g^{n q} \hat{R}_{c p d q}$, repeat=True)
    substitute(ex, $\hat{R}^{a m n b} -> -g^{a c}g^{b d}g^{m p}g^{n q} \hat{R}_{c p d q}$, repeat=True)
    substitute(ex, $\hat{R}^{m a n b} -> g^{a c}g^{b d}g^{m p}g^{n q} \hat{R}_{c p d q}$, repeat=True)
    substitute(ex, $\hat{R}^{a m}{}_{b n} -> g^{a c}g^{m p} \hat{R}_{c p b n}$, repeat=True)
    substitute(ex, $\hat{R}^{m a}{}_{b n} -> -g^{a c}g^{m p} \hat{R}_{c p b n}$, repeat=True)
    substitute(ex, $\hat{R}^{a m}{}_{n b} -> -g^{a c}g^{m p} \hat{R}_{c p b n}$, repeat=True)
    substitute(ex, $\hat{R}^{m a}{}_{n b} -> g^{a c}g^{m p} \hat{R}_{c p b n}$, repeat=True)
    
    distribute(ex)
    return ex
```

```cadabra2
def sub12d(ex):
    substitute(ex, $g^{a b} -> \delta_{a 1} \delta_{b 2} + \delta_{a 2} \delta_{b 1}$, repeat=True)
    substitute(ex, $\hat{R}_{a m b n} -> K_{m n} \delta_{a 1} \delta_{b 1} + \bar{K}_{m n} \delta_{a 2} \delta_{b 2}$, repeat=True)
    distribute(ex)
    eliminate_kronecker(ex)
    simplify(ex)

    # try some further simplifications
    substitute(ex, $g^{m n} -> \delta^{m n}$, repeat=True)
    eliminate_kronecker(ex)
    simplify(ex)

    return ex
```

```cadabra2
def pipeline(ex):
    truncate(ex)
    simplify(ex)
    lower_internal(ex)
    sub12d(ex)
    return ex
```
