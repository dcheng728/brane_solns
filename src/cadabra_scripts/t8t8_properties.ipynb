We use S= DP with rank 4, K = DP with rank2, P rank1

It is known that if rnk-4 tensor S has the same symmetries as a Riemann tensor, then

$$
\begin{align}
t_8 t_8 R^2 S^2 &= 4 R_{abcd} R_{abcd} S_{efgh} S_{efgh}
+ 16 R_{abcd} R_{abef} S_{cdgh} S_{efgh} \\
&\quad + 8 R_{abcd} R_{efgh} S_{abcd} S_{efgh}
+ 8 R_{abcd} R_{efgh} S_{abef} S_{cdgh} \\
&\quad - 64 R_{abcd} R_{abce} S_{dfgh} S_{efgh}
- 32 R_{abcd} R_{abef} S_{cegh} S_{dfgh} \\
&\quad - 64 R_{abcd} R_{aefg} S_{bhfg} S_{cdeh}
- 64 R_{abcd} R_{efgh} S_{abce} S_{dfgh} \\
&\quad - 64 R_{abcd} R_{aefg} S_{bhcd} S_{ehfg}
+ 128 R_{abcd} R_{aecf} S_{bgdh} S_{egfh} \\
&\quad + 256 R_{abcd} R_{aefg} S_{bhcf} S_{dgeh}
+ 128 R_{abcd} R_{aecf} S_{bgfh} S_{dheg} \\
&\quad + 64 R_{abcd} R_{efgh} S_{aecg} S_{bfdh}, 
\label{eq:t8R2S2}
\end{align}
$$

In particular, for $t_8 t_8 R^4$, we have

$$
\begin{align}
t_8 t_8 R^4 &= 3 \times 2^7 \Biggl\{
R_{abef} R^{fg}{}_{cd} R^{dahe} R_{gh}{}^{bc}
+ \frac{1}{2} R_{abef} R^{fgbc} R_{cdgh} R^{heda} \notag \\
&\quad - \frac{1}{2} R_{abef} R^{efbc} R_{cdgh} R^{dagh}
- \frac{1}{4} R_{abef} R^{ef}{}_{cd} R^{dagh} R_{gh}{}^{bc} \notag \\
&\quad + \frac{1}{16} R_{abef} R^{abgh} R_{ghcd} R^{cdef}
+ \frac{1}{32} (R_{abcd} R^{abcd})^2
\Biggr\}. \label{eq:A2}
\end{align}
$$

```cadabra2
def sort(ex):
    sort_product(ex)
    sort_sum(ex)
    rename_dummies(ex)
    collect_terms(ex)
    return ex
    
def cleanup(ex, meld_ = False):
    distribute(ex)
    eliminate_kronecker(ex, repeat=True)
    rename_dummies(ex)
    canonicalise(ex)
    if meld_ == True: 
    # meld potentially has a bug where it's setting too many terms to zero, 
    # one has to be careful whether to use it or nut
        try:
            meld(ex)
        except RuntimeError:
            pass
        collect_terms(ex)
    return ex
```

# Useful tensor relations

- Let us define $K^{mnrs} \equiv \bar{K}^{mn}K^{rs}$, for any kinds of contraction between $K^{mnrs} T_{mnrs}$, if $T$ is real, then $K^{m n r s} \cong K^{rs mn}$. Proof: complex jugate.
    - It follows that $K^{m n r s}R_{m}{}^{t}{}_{n}{}^{u} R_{r t s u}=0$. Proof: bianchi on second R.
    - $K^{m n r s} R_{m r}{}^{t u} R_{n s t u} = 2 K^{m n r s} R_{m}{}^{t}{}_{r}{}^{u} R_{n t s u}$. Proof: bianchi.
