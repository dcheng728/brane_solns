What's interesting is that in 10d $P\bar{P}$ is two point, but in 12d they may just be a single $\hat{R}_{MNPQ}$ component

```python
import sympy as sp

m,n,r,s = sp.symbols('m n r s', real=True)
tau1 = sp.Function('tau_1', real=True)(m, n, r, s)
tau2 = sp.Function('tau_2', real=True, positive=True)(m, n, r, s)
tau = tau1 + sp.I*tau2
taub = tau1 - sp.I*tau2
```

```python
P_m = sp.Rational(1, 2) * sp.I / tau2 * sp.diff(tau, m)
Pb_n = -sp.Rational(1, 2) * sp.I / tau2 * sp.diff(taub, n)

P_m_Pb_n = P_m * Pb_n

# Exract the real part of P_m_Pb_n
P_m_Pb_n = sp.re(sp.expand(P_m_Pb_n))
P_m_Pb_n = P_m_Pb_n.replace(lambda e: e.func == sp.re, lambda e: e.args[0])
P_m_Pb_n = P_m_Pb_n.replace(lambda e: e.func == sp.im, lambda e: sp.S.Zero)
P_m_Pb_n = sp.expand(sp.simplify(P_m_Pb_n))
```
